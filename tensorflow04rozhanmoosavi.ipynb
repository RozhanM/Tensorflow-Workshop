```python
import numpy as np
import tensorflow as tf
from keras import regularizers
from keras.optimizers import Adam
from keras.models import Sequential
from keras.layers import Dense
```

```python
x = np.random.random((200,1))*100
y = (x**2) + 2*x +53
```

```python
tf.random.set_seed(30)
```

```python
model = Sequential()
model.add(Dense(20, input_shape=(1,), activation = 'selu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dense(20, activation = 'selu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dense(20, activation = 'selu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dense(1))

model.compile(optimizer = Adam(lr = 0.03),
             loss = 'mse')

model.fit(x, y, epochs=1000,
         validation_split=0.3,
         batch_size = 250)
```
